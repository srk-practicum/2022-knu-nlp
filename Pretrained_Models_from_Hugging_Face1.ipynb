```python
from transformers import pipeline
```

```python
classifier = pipeline('sentiment-analysis', model = "finiteautomata/bertweet-base-sentiment-analysis")
```

```python
test1 = 'Good day today'
classifier(test1)
```

```python
test2 = 'Are ypu ok?'
classifier(test2)
```

```python
test3 = 'OMG Stop it!!'
classifier(test3)
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
model_name = "finiteautomata/bertweet-base-sentiment-analysis"
model = TFAutoModelForSequenceClassification.from_pretrained(model_name, from_pt=True)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
inputs = tokenizer([test1, test2, test3])
```

```python
inputs
```

```python
inputs_with_padding = tokenizer([test1, test2, test3], padding = True, truncation = True, max_length = 256, return_tensors="tf")
```

```python
inputs_with_padding
```

```python
outputs = model(inputs_with_padding)
```

```python
outputs
```

```python
import tensorflow as tf
predictions = tf.nn.softmax(outputs[0], axis=-1)
```
